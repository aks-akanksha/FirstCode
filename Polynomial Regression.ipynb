# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing datasets

```python
dataset = pd.read_csv("/home/akanksha/Downloads/Position_Salaries.csv")
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,2].values
```

```python
print(x)
```

```python
print(y)
```

# Training a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

# Training a Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)
x_poly = poly_reg.fit_transform(x) #It creates a matrix of features containing x an x^2. It does not train the model.

#Now we shall train a Polynomial (Multiple)Linear Regression model with features x and x^2.
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

# Visualizing Linear regression Model

```python
plt.scatter(x,y,color = "red")
plt.plot(x, lin_reg.predict(x), color = "blue")
plt.title("Salary vs Position")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Visualizing Polynomial Regression Model

```python
plt.scatter(x,y,color = "red")
plt.plot(x, lin_reg_2.predict(x_poly), color = "blue")
plt.title("Salary vs Position")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
